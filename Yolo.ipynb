```python
import cv2
from ultralytics import YOLO
```

```python

model = YOLO("yolov8m.pt")
print(model.names)
```

```python
model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture("https://github.com/HalyshAnton/IT-Step-Pyton-AI/raw/refs/heads/main/module5/images/Highway%20driving%20during%20the%20day%20front%20view%20camera%20driving%20plate%204k%20footage.mp4")

frames = []
fig = plt.figure()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Детекція
    results = model(frame, conf=0.15, iou=0.5, imgsz=640, max_det=50, classes = [2, 3, 5, 7, 9, 11])[0]

    annotated_frame = results.plot()

    annotated_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)

    im = plt.imshow(annotated_frame, animated=True)
    frames.append([im])

cap.release()

ani = animation.ArtistAnimation(fig, frames, interval=100, blit=True, repeat_delay=1000)
ani.save("output_with_boxes.mp4")
plt.show()
```
