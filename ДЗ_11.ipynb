Ваша задача методом градієнтного спуску знайти коефіцієнти для задачі регресії. Обираємо метод лінійної регресії та дані про діаманти

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module3/data/diamonds.csv",
                 index_col=0)

df.head()
```

```python
import numpy as np
import torch
from torch.optim import SGD, Adam

# розділіть дані на X, y
y = df.price
x = df.carat

# переведіть дані в масив numpy
y = np.array(y)
x = np.array(x)

# створіть тензори без градієнтів, тип даних float32
y = torch.tensor(y, dtype=torch.float32)
x = torch.tensor(x, dtype=torch.float32)

# створіть тензори з градієнтом, оберіть довільне початкове значення, тип даних float32
coef = torch.tensor(1.0, requires_grad = True)
intercept = torch.tensor(1.0, requires_grad = True)

# вкажіть навчальний крок
lr = 0.1

# визначення оптимізатора
optimazer = Adam([coef, intercept],
                lr=lr)

# основний цикл
for t in range(120_000):
  # обрахуйте прогноз моделі за формулою лінійної регресії
  y_train = coef*x + intercept
  # обрахуйте метрику середньо квадратичного
  loss = torch.mean((y-y_train)**2)

  # не використовуйте sklearn, лише функції torch
  # див torch.mean

  # виведіть значення метрики для кожного сотого кроку
  if t%10_000==0:
    print(f"Loss - {loss}, coef: {coef}, intercept: {intercept}.")
  # обрахуйте градієнти по метриці
  optimazer.zero_grad()
  loss.backward()
  # градієнтний спуск
  optimazer.step()
```
